```python
import requests

url = "https://ptt-discussion.tw/bbs/"
header = {
    "user_agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/130.0.0.0 Safari/537.36"
}
```

```python
# 回應原始碼的字串
response = requests.get(url, headers=header)
```

```python
# 確認是否有回傳成功: 200為成功 401失敗
response
```

```python
# 回傳html
print(response.text)
```

```python
# 找 a#logo
from bs4 import BeautifulSoup

print(type(response.text))
```

```python
# 轉化html字串到beautifulsoup 物件
soup = BeautifulSoup(response.text)
# 確認物件種類
print(type(soup))
```

```python
"""
lacate tags using BeautifulSoup
soup.find()   找淺層
    -soup.find(tag_name, attribute)
    -soup.find(tag_name, dict_of_attribute)
soup.select() 可以找很多層
    -soup,select(selector) return list
"""
# 找a標籤 a#logo 回傳一個字串
soup.find("a", id="logo")
soup.find("a", {"id":"logo"})
# 找多個標籤 回傳一個list
soup.find_all("a", id="logo")
soup.find_all("a")

# select()找標籤物件
soup.select("a#logo")
soup.select('a[id="logo"]')
soup.select_one("a#logo")
soup.select_one('a[id="logo"]')
```

```python
logo_tag = soup.select("a#logo")[0]
# logo_tag
# 取得內容
print(logo_tag.text)

# 取得url
print("https://ptt-discussion.tw" + logo_tag["href"])
```
